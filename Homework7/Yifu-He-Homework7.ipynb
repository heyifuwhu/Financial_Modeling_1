# Question 1



Calculate rt-rt*

```python
rt_uk=(df_uk.iloc[:,1]-df_uk.iloc[:,2])/1200
rt_eu=(df_eu.iloc[:,1]-df_eu.iloc[:,2])/1200
rt_jp=(df_jp.iloc[:,1]-df_jp.iloc[:,2])/1200
rt_sz=(df_sz.iloc[:,1]-df_sz.iloc[:,2])/1200
```

Calculate st

```python
st_uk=[]
st_jp=[]
for i in range(8917):
    st_uk.append(np.log(df_uk.iloc[i+21,3])-np.log(df_uk.iloc[i,3]))
    st_jp.append(np.log(df_jp.iloc[i+21,3])-np.log(df_jp.iloc[i,3]))
    
```

```python
st_eu=[]
st_sz=[]
for j in range(5525):
    st_eu.append(np.log(df_eu.iloc[j+21,3])-np.log(df_eu.iloc[j,3]))
for m in range(1281):
    st_sz.append(np.log(df_sz.iloc[m+21,3])-np.log(df_sz.iloc[m,3]))
```

```python
st_uk=np.array(st_uk)
st_jp=np.array(st_jp)
st_eu=np.array(st_eu)
st_sz=np.array(st_sz)
```

Calculate excess return

```python
excess_uk=[]
for i in range(8917):
    if rt_uk[i]>=0:
        excess_uk.append(rt_uk[i]-st_uk[i])
    else:
        excess_uk.append(st_uk[i]-rt_uk[i])
        
```

```python
excess_jp=[]
for i in range(8917):
    if rt_jp[i]>=0:
        excess_jp.append(rt_jp[i]-st_jp[i])
    else:
        excess_jp.append(st_jp[i]-rt_jp[i])
```

```python
excess_eu=[]
for i in range(5525):
    if rt_eu[i]>=0:
        excess_eu.append(rt_eu[i]-st_eu[i])
    else:
        excess_eu.append(st_eu[i]-rt_eu[i])    
```

```python
excess_sz=[]
for i in range(1281):
    if rt_sz[i]>=0:
        excess_sz.append(rt_sz[i]-st_sz[i])
    else:
        excess_sz.append(st_sz[i]-rt_sz[i])    
```

```python
excess_uk=np.array(excess_uk)
excess_jp=np.array(excess_jp)
excess_eu=np.array(excess_eu)
excess_sz=np.array(excess_sz)
```

```python
t_uk=excess_uk.mean()/excess_uk.std()*np.sqrt(8938)
t_jp=excess_jp.mean()/excess_jp.std()*np.sqrt(8938)
t_sz=excess_sz.mean()/excess_sz.std()*np.sqrt(1302)
t_eu=excess_eu.mean()/excess_eu.std()*np.sqrt(5546)
print(t_uk)
print(t_jp)
print(t_sz)
print(t_eu)
```

From the t test value, they all make a significant profit

# Question 3

```python
# defind function
from sklearn.linear_model import LinearRegression
def q3(x,y):
    length =np.min([x.shape[0],y.shape[0]])
    x=x[-length:]
    y=y[-length:]
       
    Y = []
    for i in range(252,len(x)):
        
        lm  = LinearRegression().fit(x[:i,].reshape(-1,1),y[:i,].reshape(-1,1))
        yhead = lm.predict(x[i].reshape(-1,1))
        
        if yhead[0,0]> x[i]:
            Y.append(y[i,]-x[i,])
        else:
            Y.append(-y[i,]+x[i,])
    return Y
```

```python
eu = np.array(q3(rt_eu.values,st_eu))
jp = np.array(q3(rt_jp.values,st_jp))
sz = np.array(q3(rt_sz.values,st_sz))
uk = np.array(q3(rt_uk.values,st_uk))
```

```python
# defind function
from sklearn.linear_model import LinearRegression
def q4(x,y):
    length =np.min([x.shape[0],y.shape[0]])
    x=x[-length:]
    y=y[-length:]
       
    Y = []
    for i in range(len(x)-252):
        
        lm  = LinearRegression().fit(x[i:i+251,].reshape(-1,1),y[i:i+251,].reshape(-1,1))
        yhead = lm.predict(x[i].reshape(-1,1))
        
        if yhead[0,0]> x[i]:
            Y.append(y[i,]-x[i,])
        else:
            Y.append(-y[i,]+x[i,])
    return Y
```
